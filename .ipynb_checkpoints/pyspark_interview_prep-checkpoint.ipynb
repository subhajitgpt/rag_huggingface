## 9) Performance Basics (what interviewers really care about)
Key terms to be able to explain:
- **Partition**: how data is split across tasks
- **Shuffle**: data movement across partitions (expensive)
- **Cache/Persist**: keep computed data for reuse
- **Broadcast join**: replicate small table to avoid shuffle
- **Predicate pushdown / column pruning**: read less data
- **Skew**: uneven data distribution causing slow tasks
