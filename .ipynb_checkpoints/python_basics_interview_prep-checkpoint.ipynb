## 1) Variables, Types, and Mutability
In interviews, the *mutability* distinction matters:
- **Immutable**: `int`, `float`, `bool`, `str`, `tuple`, `frozenset`
- **Mutable**: `list`, `dict`, `set`

Mutability affects copying, function arguments, and bugs.

```python
# Variables are references to objects (names bound to objects)
x = 10
y = x  # y refers to the same int object as x (ints are immutable)
x = x + 1  # creates a NEW int object; x now refers to it
print('x:', x, 'y:', y)

# Mutable example: lists can change in-place
a = [1, 2, 3]
b = a  # b refers to same list object
a.append(4)  # in-place mutation affects b too
print('a:', a, 'b:', b)
```

### Common gotcha: default mutable arguments
A classic interview question: why is this a bug?

```python
def add_item_bad(item, bucket=[]):
    # BAD: bucket is created once at function definition time
    bucket.append(item)
    return bucket

print(add_item_bad('a'))
print(add_item_bad('b'))  # surprises many people

def add_item_good(item, bucket=None):
    # GOOD: create a new list per call unless caller provides one
    if bucket is None:
        bucket = []
    bucket.append(item)
    return bucket

print(add_item_good('a'))
print(add_item_good('b'))
```

## 2) Numbers, Booleans, Operators
Know the difference between:
- `==` (value equality) vs `is` (identity / same object)
- integer division `//` vs true division `/`
- short-circuit: `and`, `or`

```python
a = 256
b = 256
print(a == b)  # value equal
print(a is b)  # implementation detail: small ints may be interned

print(7 / 2)   # 3.5
print(7 // 2)  # 3
print(7 % 2)   # 1

# Short-circuiting can avoid errors
s = ''
print(s and s[0])  # does not crash because s is falsy
```

## 3) Strings
Frequently tested:
- slicing (`s[a:b:c]`)
- immutability
- `split`, `join`, `strip`, `replace`
- f-strings

```python
s = '  hello world  '
print(s.strip())
print(s.upper())
print('hello' in s)

t = 'abcd'
print(t[1:3])   # 'bc'
print(t[::-1])  # reverse

parts = ['2026', '01', '16']
print('-'.join(parts))

name = 'Ada'
age = 36
print(f'{name} is {age} years old')
```

## 4) Core Collections: list, tuple, set, dict
Interview focus:
- Time complexity intuition (average-case)
  - `x in list` is $O(n)$
  - `x in set` / `x in dict` is $O(1)$
- When to use each structure

```python
nums = [3, 1, 2, 3]
print(nums.count(3))
nums.sort()  # in-place
print(nums)

# Tuples are immutable: good for fixed records, dict keys, etc.
point = (10, 20)
x, y = point  # unpacking
print(x, y)

# Sets deduplicate and support fast membership
unique = set([1, 2, 2, 3])
print(unique)

# Dicts map keys -> values
ages = {'Ada': 36, 'Alan': 41}
ages['Grace'] = 50
print(ages.get('Missing', -1))
print(list(ages.keys()))
print(list(ages.values()))
```

### Copying: shallow vs deep
Shallow copy duplicates the outer container, not nested objects.
Deep copy duplicates everything (can be expensive).

```python
import copy

matrix = [[1, 2], [3, 4]]
shallow = list(matrix)      # or matrix.copy()
deep = copy.deepcopy(matrix)

matrix[0].append(99)
print('matrix:', matrix)
print('shallow:', shallow)  # shares inner lists
print('deep:', deep)        # independent
```

## 5) Control Flow
Be comfortable with `if/elif/else`, loops, `break/continue`, and `for/else`.

```python
# for/else: else runs only if the loop DID NOT break
target = 7
data = [1, 3, 5, 7, 9]

for n in data:
    if n == target:
        print('found', target)
        break
else:
    print('not found')
```

## 6) Functions
Know:
- positional vs keyword arguments
- default arguments
- `*args` and `**kwargs`
- scope (LEGB)
- docstrings

```python
def greet(name, greeting='Hello', *, punctuation='!'):
    """Return a greeting string."""
    # Note the *: forces punctuation to be keyword-only
    return f'{greeting}, {name}{punctuation}'

print(greet('Ada'))
print(greet('Ada', 'Hi', punctuation='.'))

def summarize(*args, **kwargs):
    # args: tuple of extra positional args
    # kwargs: dict of extra keyword args
    return {'args': args, 'kwargs': kwargs}

print(summarize(1, 2, 3, lang='python', level='basic'))
```

### Closures + LEGB (Local, Enclosing, Global, Built-in)
Closures are common in interviews, especially with callbacks and decorators.

```python
def make_multiplier(factor):
    # factor is captured by the inner function (closure)
    def mul(x):
        return x * factor
    return mul

times3 = make_multiplier(3)
print(times3(10))  # 30
```

## 7) Comprehensions
List/set/dict comprehensions are frequently tested. Prioritize readability.

```python
nums = [1, 2, 3, 4, 5, 6]
evens = [n for n in nums if n % 2 == 0]
squares = {n: n*n for n in nums}
unique_lengths = {len(word) for word in ['a', 'aa', 'bbb', 'aa']}
print(evens)
print(squares)
print(unique_lengths)
```

## 8) Exceptions
Know:
- catching specific exceptions
- `try/except/else/finally`
- raising your own errors

```python
def parse_int(text):
    try:
        value = int(text)
    except ValueError as e:
        # In interviews: explain why you catch ValueError (not Exception)
        raise ValueError(f'Not an int: {text!r}') from e
    else:
        # Runs if no exception occurred
        return value
    finally:
        # Runs no matter what (cleanup), avoid heavy logic here
        pass

for item in ['42', 'x']:
    try:
        print(item, '->', parse_int(item))
    except ValueError as e:
        print('error:', e)
```

## 9) Iteration: iterators, generators
Generators are memory-efficient and show up often in interviews.

```python
# Generator function yields values lazily
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for x in countdown(5):
    print(x)

# Generator expression
gen = (i*i for i in range(5))
print(list(gen))
```

## 10) Decorators (basics)
Decorators wrap functions. Interviewers often ask how `@decorator` works.

```python
import time
from functools import wraps

def timed(func):
    # wraps preserves __name__ and docstring
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        elapsed = time.perf_counter() - start
        print(f'{func.__name__} took {elapsed:.6f}s')
        return result
    return wrapper

@timed
def slow_add(a, b):
    time.sleep(0.05)
    return a + b

print(slow_add(2, 3))
```

## 11) Files + Context Managers
Prefer `with open(...)` so the file closes even on errors.

```python
# Demonstration using a temporary file (safe for notebooks)
from pathlib import Path

tmp_path = Path('tmp_interview_demo.txt')

with tmp_path.open('w', encoding='utf-8') as f:
    f.write('line1\nline2\n')

with tmp_path.open('r', encoding='utf-8') as f:
    for line in f:
        print('read:', line.rstrip())

# Cleanup
tmp_path.unlink(missing_ok=True)
```

## 12) Object-Oriented Programming (OOP)
Know:
- class vs instance attributes
- `__init__`
- `@property`
- inheritance basics

```python
class BankAccount:
    # class attribute shared by all instances
    bank_name = 'Interview Bank'

    def __init__(self, owner, balance=0):
        self.owner = owner        # instance attribute
        self._balance = balance   # leading underscore: internal

    @property
    def balance(self):
        return self._balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError('amount must be positive')
        self._balance += amount

    def __repr__(self):
        return f'BankAccount(owner={self.owner!r}, balance={self._balance})'

acct = BankAccount('Ada', 100)
acct.deposit(50)
print(acct)
print(BankAccount.bank_name, acct.bank_name)
```

## 13) Dataclasses (handy for interviews)
Dataclasses reduce boilerplate for simple data-holding classes.

```python
from dataclasses import dataclass

@dataclass(frozen=True)
class Point:
    x: int
    y: int

p = Point(1, 2)
print(p)
# p.x = 99  # would raise: frozen=True makes it immutable
```

## 14) Common Interview Exercises (practice)
Try implementing these quickly, then discuss complexity and edge cases:
1. Reverse a string
2. Check palindrome
3. Two-sum
4. Frequency count
5. Validate parentheses

```python
# 1) Reverse a string (two common ways)
def reverse_string(s: str) -> str:
    # slicing is concise and fast in CPython
    return s[::-1]

print(reverse_string('interview'))

# 2) Palindrome check
def is_palindrome(s: str) -> bool:
    # Normalize: keep it simple; interviews may ask to ignore case/spaces
    s = s.lower()
    return s == s[::-1]

print(is_palindrome('Level'))
print(is_palindrome('Python'))
```

```python
# 3) Two-sum: return indices of two numbers that add up to target
# Typical expected approach: hash map (dict) for O(n) time
def two_sum(nums, target):
    seen = {}  # value -> index
    for i, n in enumerate(nums):
        needed = target - n
        if needed in seen:
            return (seen[needed], i)
        seen[n] = i
    return None

print(two_sum([2, 7, 11, 15], 9))
print(two_sum([3, 2, 4], 6))
```

```python
# 4) Frequency count
from collections import Counter

def freq_count(items):
    # Counter is the standard library solution
    return Counter(items)

print(freq_count('banana'))

# 5) Validate parentheses (balanced brackets)
def is_valid_parentheses(s: str) -> bool:
    pairs = {')': '(', ']': '[', '}': '{'}
    stack = []
    for ch in s:
        if ch in '([{':
            stack.append(ch)
        elif ch in pairs:
            if not stack or stack[-1] != pairs[ch]:
                return False
            stack.pop()
        else:
            # If interviews include other chars, decide policy
            pass
    return not stack

tests = ['()', '([])', '(]', '([)]', '{[()]}', '']
for t in tests:
    print(t, '->', is_valid_parentheses(t))
```
