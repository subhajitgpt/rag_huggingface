# Spark SQL — Interview Prep (SQL basics)

This notebook focuses on **Spark SQL** patterns commonly asked in interviews:
- SELECT / WHERE / ORDER BY / LIMIT
- GROUP BY / HAVING
- JOINs (inner/left/right/full, semi/anti)
- Nested queries + CTEs
- Window functions
- CASE WHEN, NULL handling, common functions

Everything runs on a small in-memory dataset (no external files).

## 0) Setup Spark + sample tables
We’ll create a few temp views: `customers`, `orders`, `order_items`, `products`.
All remaining examples use `spark.sql(...)` only.

```python
try:
    from pyspark.sql import SparkSession
    from pyspark.sql import functions as F
except Exception as e:
    raise RuntimeError('PySpark not available. Install with: pip install pyspark') from e

spark = (
    SparkSession.builder
    .appName('spark-sql-interview-prep')
    .master('local[*]')
    .getOrCreate()
)
spark.sparkContext.setLogLevel('WARN')

customers = spark.createDataFrame(
    [
        (1, 'Ada',   'US'),
        (2, 'Alan',  'UK'),
        (3, 'Grace', 'US'),
        (4, 'Linus', 'FI'),
    ],
    ['customer_id', 'customer_name', 'country']
)

orders = spark.createDataFrame(
    [
        (101, 1, '2026-01-10', 'PAID'),
        (102, 2, '2026-01-11', 'PAID'),
        (103, 2, '2026-01-13', 'CANCELLED'),
        (104, 99,'2026-01-12', 'PAID'),  # customer_id=99 missing on purpose
        (105, 3, '2026-01-12', 'PAID'),
    ],
    ['order_id', 'customer_id', 'order_date', 'status']
).withColumn('order_date', F.to_date('order_date'))

products = spark.createDataFrame(
    [
        (10, 'Keyboard', 50.0),
        (11, 'Mouse',    25.0),
        (12, 'Monitor',  200.0),
        (13, 'Cable',     10.0),
    ],
    ['product_id', 'product_name', 'unit_price']
)

order_items = spark.createDataFrame(
    [
        (101, 10, 1),
        (101, 11, 2),
        (102, 12, 1),
        (103, 13, 5),
        (105, 10, 1),
        (105, 12, 2),
    ],
    ['order_id', 'product_id', 'qty']
)

customers.createOrReplaceTempView('customers')
orders.createOrReplaceTempView('orders')
products.createOrReplaceTempView('products')
order_items.createOrReplaceTempView('order_items')

spark.sql('SHOW TABLES').show(truncate=False)
```

## 1) SELECT + WHERE + ORDER BY + LIMIT
Core SQL basics: projection, filtering, sorting, limiting results.

```python
spark.sql('''
SELECT
  order_id, customer_id, order_date, status
FROM orders
WHERE status = 'PAID'
ORDER BY order_date DESC, order_id
LIMIT 10
''').show()
```

## 2) GROUP BY + aggregate functions
Know common aggregates: `COUNT`, `SUM`, `AVG`, `MIN`, `MAX`.
In interviews: clarify if `COUNT(*)` vs `COUNT(col)` matters (nulls).

```python
spark.sql('''
SELECT
  status,
  COUNT(*) AS orders,
  COUNT(DISTINCT customer_id) AS unique_customers
FROM orders
GROUP BY status
ORDER BY orders DESC
''').show()
```

## 3) HAVING (filter after aggregation)
`WHERE` filters rows **before** grouping; `HAVING` filters **after** grouping.

```python
spark.sql('''
SELECT
  customer_id,
  COUNT(*) AS total_orders
FROM orders
GROUP BY customer_id
HAVING COUNT(*) >= 2
ORDER BY total_orders DESC
''').show()
```

## 4) JOIN basics (INNER/LEFT/RIGHT/FULL)
Interview checklist:
- join keys and expected cardinality (1:1, 1:many, many:many)
- what happens to non-matching rows in each join type

```python
# INNER JOIN: only rows with matching customer_id in both tables
spark.sql('''
SELECT
  o.order_id, o.customer_id, c.customer_name, c.country, o.status
FROM orders o
JOIN customers c
  ON o.customer_id = c.customer_id
ORDER BY o.order_id
''').show()

# LEFT JOIN: keep ALL orders, even if customer is missing
spark.sql('''
SELECT
  o.order_id, o.customer_id, c.customer_name, o.status
FROM orders o
LEFT JOIN customers c
  ON o.customer_id = c.customer_id
ORDER BY o.order_id
''').show()

# FULL OUTER JOIN: keep both sides
spark.sql('''
SELECT
  o.order_id,
  o.customer_id AS order_customer_id,
  c.customer_id AS customer_customer_id,
  c.customer_name
FROM orders o
FULL OUTER JOIN customers c
  ON o.customer_id = c.customer_id
ORDER BY o.order_id
''').show()
```

## 5) Multi-table join (orders → items → products)
This is a classic interview pattern: compute order totals.

```python
spark.sql('''
SELECT
  o.order_id,
  o.customer_id,
  SUM(oi.qty * p.unit_price) AS order_total
FROM orders o
JOIN order_items oi
  ON o.order_id = oi.order_id
JOIN products p
  ON oi.product_id = p.product_id
GROUP BY o.order_id, o.customer_id
ORDER BY order_total DESC
''').show()
```

## 6) Nested query (subquery)
Common interview ask: *find customers whose order_total exceeds X*.

```python
spark.sql('''
SELECT
  t.customer_id, c.customer_name, t.total_spend
FROM (
  SELECT
    o.customer_id,
    SUM(oi.qty * p.unit_price) AS total_spend
  FROM orders o
  JOIN order_items oi ON o.order_id = oi.order_id
  JOIN products p ON oi.product_id = p.product_id
  WHERE o.status = 'PAID'
  GROUP BY o.customer_id
) t
JOIN customers c ON t.customer_id = c.customer_id
WHERE t.total_spend >= 200
ORDER BY t.total_spend DESC
''').show()
```

## 7) CTEs (`WITH ...`)
CTEs make complex queries readable and are often preferred in interviews.

```python
spark.sql('''
WITH order_totals AS (
  SELECT
    o.order_id,
    o.customer_id,
    o.status,
    SUM(oi.qty * p.unit_price) AS order_total
  FROM orders o
  JOIN order_items oi ON o.order_id = oi.order_id
  JOIN products p ON oi.product_id = p.product_id
  GROUP BY o.order_id, o.customer_id, o.status
),
paid_orders AS (
  SELECT * FROM order_totals WHERE status = 'PAID'
)
SELECT
  c.country,
  COUNT(*) AS paid_orders,
  ROUND(AVG(po.order_total), 2) AS avg_paid_order_total
FROM paid_orders po
JOIN customers c ON po.customer_id = c.customer_id
GROUP BY c.country
ORDER BY avg_paid_order_total DESC
''').show()
```

## 8) Window functions
Very common in interviews: rank, top-N per group, running totals, dedup with ordering.

```python
spark.sql('''
WITH order_totals AS (
  SELECT
    o.order_id,
    o.customer_id,
    o.order_date,
    o.status,
    SUM(oi.qty * p.unit_price) AS order_total
  FROM orders o
  JOIN order_items oi ON o.order_id = oi.order_id
  JOIN products p ON oi.product_id = p.product_id
  GROUP BY o.order_id, o.customer_id, o.order_date, o.status
)
SELECT
  *,
  ROW_NUMBER() OVER (PARTITION BY customer_id ORDER BY order_total DESC) AS rn,
  DENSE_RANK() OVER (PARTITION BY customer_id ORDER BY order_total DESC) AS dr
FROM order_totals
ORDER BY customer_id, rn
''').show()
```

## 9) CASE WHEN + NULL handling
Know `CASE WHEN`, `COALESCE`, and basic null-safe logic.

```python
spark.sql('''
SELECT
  o.order_id,
  o.customer_id,
  COALESCE(c.customer_name, 'UNKNOWN') AS customer_name,
  o.status,
  CASE
    WHEN o.status = 'PAID' THEN 'good'
    WHEN o.status = 'CANCELLED' THEN 'bad'
    ELSE 'other'
  END AS status_bucket
FROM orders o
LEFT JOIN customers c ON o.customer_id = c.customer_id
ORDER BY o.order_id
''').show()
```

## 10) Common functions (strings + dates)
Know a few everyday functions: `LOWER`, `UPPER`, `TRIM`, `SUBSTR`, `DATE_ADD`, `DATEDIFF`.

```python
spark.sql('''
SELECT
  customer_id,
  UPPER(customer_name) AS name_upper,
  LOWER(country) AS country_lower
FROM customers
ORDER BY customer_id
''').show()

spark.sql('''
SELECT
  order_id,
  order_date,
  DATE_ADD(order_date, 7) AS plus_7_days
FROM orders
ORDER BY order_id
''').show()
```

## 11) UNION / UNION ALL
- `UNION` removes duplicates (like `DISTINCT`)
- `UNION ALL` keeps duplicates (faster)

```python
spark.sql('''
SELECT customer_id FROM orders WHERE status = 'PAID'
UNION ALL
SELECT customer_id FROM orders WHERE status = 'CANCELLED'
''').show()
```

## 12) EXPLAIN (query plan)
In Spark interviews, it’s great to mention `EXPLAIN` to debug performance (joins/shuffles).

```python
spark.sql('''
EXPLAIN FORMATTED
SELECT
  c.country, SUM(oi.qty * p.unit_price) AS total_spend
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
WHERE o.status = 'PAID'
GROUP BY c.country
''').show(truncate=False)
```
