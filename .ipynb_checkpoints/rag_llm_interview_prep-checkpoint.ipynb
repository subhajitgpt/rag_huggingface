# RAG (Retrieval-Augmented Generation) + LLM Integrations

This notebook is a practical starter for **RAG** and **LLM** integrations using:
- **OpenAI** (API)
- **Hugging Face** (local `transformers` pipeline)
- **LLaMA** local inference via **llama-cpp-python** (GGUF models)

Everything is heavily commented and designed for interview prep + demos.

## What is RAG?
RAG = retrieve relevant context (documents/chunks) → inject into prompt → generate answer.

### Why interviewers like RAG
- Shows you can reduce hallucinations by grounding answers in context
- Demonstrates pipeline thinking: ingestion → chunking → retrieval → generation → evaluation

## 0) Dependencies and setup
This notebook can run in multiple modes:
- **RAG retrieval only** (no LLM): uses `scikit-learn` TF-IDF (lightweight).
- **OpenAI**: requires `openai` package + `OPENAI_API_KEY` env var.
- **Hugging Face local**: requires `transformers` + a backend (usually `torch`).
- **LLaMA local**: requires `llama-cpp-python` + a local `.gguf` model file.

You can start with retrieval-only, then turn on whichever LLM backend you prefer.

```python
import os
import textwrap
from dataclasses import dataclass

import numpy as np

# Lightweight retrieval using TF-IDF (available via scikit-learn)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

np.random.seed(42)
```

## 1) A tiny document corpus (for demo)
In real projects, documents come from PDFs, web pages, tickets, DB rows, etc.
Here we keep it in-memory so you can run this anywhere.

```python
documents = [
    {
        'id': 'doc-1',
        'title': 'Refund policy',
        'text': 'Refunds are available within 30 days of purchase with a valid receipt. Digital goods are non-refundable once downloaded.',
    },
    {
        'id': 'doc-2',
        'title': 'Shipping policy',
        'text': 'Standard shipping takes 3-5 business days. Expedited shipping takes 1-2 business days. International shipping may take 10-15 days.',
    },
    {
        'id': 'doc-3',
        'title': 'Account security',
        'text': 'Enable two-factor authentication (2FA) to secure your account. Passwords must be at least 12 characters. Do not reuse passwords.',
    },
    {
        'id': 'doc-4',
        'title': 'Troubleshooting login',
        'text': 'If you cannot log in, reset your password. Check for caps lock. If 2FA fails, ensure your phone time is correct or use backup codes.',
    },
]

doc_texts = [f
 for d in documents]
len(doc_texts), doc_texts[0][:80]
```

## 2) Retrieval with TF-IDF (baseline RAG retrieval)
This is NOT the only way to do retrieval (embeddings are common), but it’s a good lightweight baseline for interviews.

Steps:
1. Fit TF-IDF vectorizer on documents
2. Vectorize query
3. Compute cosine similarity
4. Pick top-k documents

```python
vectorizer = TfidfVectorizer(stop_words='english')
doc_matrix = vectorizer.fit_transform(doc_texts)

def retrieve_top_k(query: str, k: int = 2):
    q_vec = vectorizer.transform([query])
    sims = cosine_similarity(q_vec, doc_matrix).ravel()
    top_idx = np.argsort(-sims)[:k]
    results = []
    for i in top_idx:
        results.append({
            'score': float(sims[i]),
            'doc': documents[i],
        })
    return results

query = 'How long do international deliveries take?'
hits = retrieve_top_k(query, k=2)
for h in hits:
    print(round(h['score'], 3), '-', h['doc']['title'])
```

## 4) LLM backend abstraction
In real systems, you want to swap LLM providers without rewriting your RAG logic.
We'll create a tiny interface and three implementations.

```python
class LLMClient:
    def generate(self, prompt: str) -> str:
        raise NotImplementedError


@dataclass
class OpenAIChatClient(LLMClient):
    model: str = 'gpt-4o-mini'

    def generate(self, prompt: str) -> str:
        # Uses the modern OpenAI Python SDK (openai>=1.x)
        # Auth reads from env var OPENAI_API_KEY
        try:
            from openai import OpenAI
        except Exception as e:
            raise RuntimeError('Install OpenAI SDK: pip install openai') from e

        if not os.getenv('OPENAI_API_KEY'):
            raise RuntimeError('Set env var OPENAI_API_KEY before calling OpenAI.')

        client = OpenAI()
        resp = client.chat.completions.create(
            model=self.model,
            messages=[
                {'role': 'system', 'content': 'You answer using provided context only.'},
                {'role': 'user', 'content': prompt},
            ],
            temperature=0.2,
        )
        return resp.choices[0].message.content


@dataclass
class HuggingFaceLocalClient(LLMClient):
    # Small default model for demo; not instruction-tuned
    model_id: str = 'distilgpt2'

    def generate(self, prompt: str) -> str:
        # Local generation via transformers pipeline
        # Requires: pip install transformers torch
        try:
            from transformers import pipeline
        except Exception as e:
            raise RuntimeError('Install: pip install transformers torch') from e

        gen = pipeline('text-generation', model=self.model_id)
        out = gen(prompt, max_new_tokens=120, do_sample=True, temperature=0.7)[0]['generated_text']
        # The pipeline returns prompt + completion; keep it simple
        return out


@dataclass
class LlamaCppClient(LLMClient):
    # Requires a local GGUF model file. Example: Llama-3.2-1B-Instruct.Q4_K_M.gguf
    gguf_path: str

    def generate(self, prompt: str) -> str:
        # Local inference via llama-cpp-python
        # Install: pip install llama-cpp-python
        try:
            from llama_cpp import Llama
        except Exception as e:
            raise RuntimeError('Install: pip install llama-cpp-python') from e

        if not os.path.exists(self.gguf_path):
            raise RuntimeError(f'GGUF file not found: {self.gguf_path}')

        llm = Llama(
            model_path=self.gguf_path,
            n_ctx=2048,
            n_threads=max(1, os.cpu_count() or 1),
        )

        # llama-cpp completions are plain-text (unless you use a chat template)
        res = llm(prompt, max_tokens=200, temperature=0.2)
        return res['choices'][0]['text']
```

## 5) Run the RAG pipeline end-to-end
Choose an LLM backend by instantiating one client.

Examples:
- `client = OpenAIChatClient(model='gpt-4o-mini')`
- `client = HuggingFaceLocalClient(model_id='distilgpt2')`
- `client = LlamaCppClient(gguf_path='PATH/TO/MODEL.gguf')`

```python
def rag_answer(question: str, client: LLMClient, k: int = 2) -> str:
    contexts = retrieve_top_k(question, k=k)
    prompt = build_rag_prompt(question, contexts)
    return client.generate(prompt)

# --- Pick ONE backend below ---
# client = OpenAIChatClient(model='gpt-4o-mini')
# client = HuggingFaceLocalClient(model_id='distilgpt2')
# client = LlamaCppClient(gguf_path=r'C:\path\to\model.gguf')

# For now, demonstrate retrieval + prompt without calling any model:
q = 'What is the refund window?'
ctx = retrieve_top_k(q, k=2)
print(build_rag_prompt(q, ctx))
```

## 9) Interview notes (talk track)
If asked to design a production RAG system, mention:
- Chunking strategy (size/overlap)
- Embeddings + vector database (FAISS, Pinecone, pgvector, etc.)
- Filtering + metadata (tenant, date, access control)
- Reranking (optional)
- Prompting + citations
- Evaluation (faithfulness, answer correctness, retrieval recall)
- Cost + latency tradeoffs (cache, batch, smaller models)
- Safety: avoid leaking secrets, PII handling, prompt injection defenses
